```python
import os
import datetime
from pathlib import Path
import shutil
```

```python
file_types = {"Images": [".jpg", ".jpeg", ".png", ".gif"],
    "Documents": [".pdf", ".docx", ".txt"],
    "Videos": [".mp4", ".mov"],
    "Audio": [".mp3", ".wav"],}
```

```python
def organize_files_type(path):
    target_folder = Path(path)
    for file in target_folder.iterdir():
        if file.is_file():
            extension = file.suffix.lower()
            destination = None
            for types, exts in file_types.items():
                if extension in exts:
                    destination = target_folder / types
                    break
            if destination is None:
                destination = target_folder / "Misc."
            destination.mkdir(exist_ok = True)
            file.rename(destination / file.name)
    print("The folder has been organized")
```

```python
organize_files_type("Sources")
```
