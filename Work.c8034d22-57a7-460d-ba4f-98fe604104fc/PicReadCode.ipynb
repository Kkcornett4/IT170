```python
from PIL import Image
import pytesseract as tess
##Pytesseract library imports --> pip install pytesseract
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
##ReportLab library imports --> pip install reportlab

tess.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
```

```python
import os

#Function that completes the process of writing the image.
#Open image through pytesseract
def writeImage(image):
    text = tess.image_to_string(Image.open(image))

# Get image filename
    file_root = os.path.splitext(os.path.basename(image))[0]
    new_filename = f"{file_root}.txt"

    outputLoc = "Outputs"
    pdf_filename = os.path.join(outputLoc, f"{file_root}.pdf")


# Create new txt file using string and filename
    with open(new_filename, "w", encoding = "utf-8", errors = "replace") as f_txt:
        f_txt.write(text)

    c = canvas.Canvas(pdf_filename, pagesize = letter)
    width, height = letter
    
    textobject = c.beginText(40, height - 40)
    textobject.setFont("Helvetica", 12)
    
    with open(new_filename, "r", encoding = "utf-8", errors = "replace") as f_txt:
        for line in f_txt:
            textobject.textLine(line.rstrip())
    
    c.drawText(textobject)
    c.save()
```

```python
writeImage("Example image.png")
```

```python
writeImage("Decolonizing_notes.jpg")
```
