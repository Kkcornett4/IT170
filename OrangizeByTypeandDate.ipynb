```python
import os
import datetime
from pathlib import Path
import shutil
```

```python
file_types = {"Images": [".jpg", ".jpeg", ".png", ".gif"],
    "Documents": [".pdf", ".docx", ".txt"],
    "Videos": [".mp4", ".mov"],
    "Audio": [".mp3", ".wav"],}
```

```python
def organize_files_type_and_date(path):
    target_folder = Path(path)
    for file in target_folder.iterdir():
        if file.is_file():
            extension = file.suffix.lower()
            file_type = "Misc"
            for types, exts in file_types.items():
                if extension in exts:
                    file_type = types
                    break
            timestamp = file.stat().st_mtime
            date_str = datetime.fromtimestamp(timestamp).strftime("%Y-%m-%d")
            destination = target_folder/file_type/date_str
            destination.mkdir(parents=True, exist_ok=True)
            file.rename(destination/file.name)
    print("The folder has been organized.")
```

```python
organize_files_type_and_date("Sources")
```
